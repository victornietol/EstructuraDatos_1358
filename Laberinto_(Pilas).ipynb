```python
class StackADT:
  def __init__(self):
    self.data = list()
    self.tope = 0

  def is_empty(self):
    return self.tope == 0

  def length(self):
    return self.tope 

  def pop(self):
    if not self.is_empty():    # en otros lenguajes "not" es igual "!"
      self.tope -=1
      return self.data.pop()
  
  def peek(self):
    return self.data[self.tope -1]    # tambien se puede usar --> return self.data[ self.length() - 1]

  def push(self, value):
    self.data.append(value)
    self.tope += 1

  def to_string(self):
    print("Estado de la pìla es: ")
    for elem in self.data[::-1]:
      print(f"| {elem} |")
```

```python
pila = StackADT()
pila.push(10)
pila.push(20)
pila.push(30)
pila.to_string()

print("llamada a peek", pila.peek())
print("llamada a pop", pila.pop())

pila.to_string()
pila.push(25)
pila.push(26)
print("despues de meter 25 y 26")
pila.to_string()
print(f"la pila tiene {pila.length()} elementos")
pila.pop()
pila.pop()
pila.pop()
pila.pop()
print(f"la pila tiene {pila.length()} elementos")
```

```python
class Array2D():
  def __init__(self, ren, col):
    self.__columnas = col
    self.__renglones = ren
    self.__data = [[ 0 for i in range(col)]for i in range (ren)]

  def to_string(self):
    for i in range(self.__renglones):
      print(self.__data[i])

  def clear(self, dato):
    for i in range (self.__renglones):
      for x in range(self.__columnas):
        self.__data[i][x] = dato

  def get_row_size(self):
    return self.__renglones

  def get_col_size(self):
    return self.__columnas

  def set_item(self,ren,col,dato):
    self.__data[ren][col] = dato

  def get_item(self,ren,col):
    return self.__data[ren][col]
```

```python
class LaberintoADT:
  def __init__(self, archivo):
    self.laberinto = None  # Array2D
    self.copia = None
    self.camino = StackADT()
    self.visitados = []
    self.inicializa( archivo )
    self.guardar_copia()
    self.laberinto.to_string()
    print("Stack")
    self.camino.to_string()
    self.previa = None

  """
  P -> pared
  C -> corredor
  E -> entrada
  S -> salida
  X -> Ya visitado se mete en una lista

  buscarCamino
  yaVisitado(x, y) --> True o False , los ya visitados los vamos a meter a una lista
  marcarComoVisitado(x,y) --> Para backtraking
  mostrarCamino --> Imprimir pila
  guardarCopia --> Crear otra variable en el constructor
  esSalida(x, y) --> True o False
  encontrarLimites(x, y) --> True o False
  esCorredor(x, y) --> True , False
  """
    
  def inicializa(self, archivo):
    entrada = open( archivo, 'rt')
    self.laberinto = Array2D( int(entrada.readline().strip()) , int(entrada.readline().strip()) )
    self.laberinto.clear('P')
    lineas = entrada.readlines()
    #print("Lineas: ")
    #print(lineas)
    for r in range( len(lineas) ):
      lineaBis = lineas[r].strip().split(',')
      for c in range( len(lineaBis) ):
        self.laberinto.set_item(r,c,lineaBis[c])

    # buscar la entrada
    for r in range( self.laberinto.get_row_size() ):
      for c in range( self.laberinto.get_col_size() ):
        if self.laberinto.get_item(r,c) == 'E':
          self.camino.push( (r,c) )

  def guardar_copia(self):
    self.copia = self.laberinto

  def get_copia(self):
    return self.copia

  # True cuando ya fue visitado
  def ya_visitado(self, r, c):
    resultado = False
    if (r,c) in self.visitados:  # [ (13,2) , (12,2)]
      resultado = True
    return resultado

  def marcar_visitado(self, r, c):
    self.visitados.append( (r, c) )

  def mostrar_camino(self):
    self.camino.to_string()
    # pintar el camino en el laberinto

  def es_salida(self):
    actual = self.camino.peek()
    resultado = False
    if self.laberinto.get_item(actual[0], actual[1]) == 'S':
      resultado = True
    return resultado

  # True esta dentro, False esta fuera del array
  def verificar_limites(self, r, c):
    limite_correcto = True                # \ <-- es para indicar que el codigo sigue abajo
    if r < 0 or r > self.laberinto.get_row_size() or c < 0 or c > self.laberinto.get_col_size:
      limite_correcto = False
    return limite_correcto

  # True si esa posicion hay una 'C' o 'S', False si no lo es
  def es_corredor(self, r, c):
    resultado = False
    if self.laberinto.get_item(r, c) == 'C' or self.laberinto.get_item(r, c) == 'S':
      resultado = True
    return resultado

  def set_previa(self, pos_previa):
    self.previa = pos_previa

  def get_previa(self):
    return self.previa

  def buscar_camino(self):
    actual = self.camino.peek()
    ren_actual = actual[0]
    col_actual = actual[1]
    # Izquierda c-1
    if self.verificar_limites(ren_actual, col_actual -1) and \
    self.es_corredor(ren_actual, col_actual -1) and \
    not (self.ya_visitado(ren_actual, col_actual -1)):
      self.set_previa( actual )
      self.camino.push( (ren_actual, col_actual -1) )

    # Arriba r-1
    elif self.verificar_limites(ren_actual -1, col_actual) and \
    self.es_corredor(ren_actual -1, col_actual) and \
    not (self.ya_visitado(ren_actual -1, col_actual)):
      self.set_previa( actual )
      self.camino.push( (ren_actual -1, col_actual) )

    # Derecha c+1
    elif self.verificar_limites(ren_actual, col_actual +1) and \
    self.es_corredor(ren_actual, col_actual +1) and \
    not (self.ya_visitado(ren_actual, col_actual +1)):
      self.set_previa( actual )
      self.camino.push( (ren_actual, col_actual +1) )

    # Abajo r+1
    elif self.verificar_limites(ren_actual +1, col_actual) and \
    self.es_corredor(ren_actual +1, col_actual) and \
    not (self.ya_visitado(ren_actual +1, col_actual)):
      self.set_previa( actual )
      self.camino.push( (ren_actual +1, col_actual) )

    else:
      print("Backtracking")
      self.marcar_visitado( ren_actual , col_actual )
      self.camino.pop()
      # backtracking
```
